```python
import matplotlib.pyplot as plt
import numpy as np

def array_creator(s):
    a = s.split(" ")
    y = []
    c = 0
    z = []
    for i in range(2304):
        if c == 0:
            z.append(float(a[i]))
        elif c%48 != 0:
            z.append(float(a[i]))
        else:
            y.append(z)
            z = []
            z.append(a[i])
        c += 1
    y.append(z)
    return(y)

for i in range(100,105):
    img = array_creator(file.pixels[i])
    img = np.array(img)
    img = np.reshape(img, (48, 48, -1))
    img.ndim
    img = img.astype(float)
    plt.figure()
    plt.imshow(img)
```
